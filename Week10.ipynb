```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('data.csv')

data['age'] = pd.to_numeric(data['age'], errors='coerce')

data['renta'] = pd.to_numeric(data['renta'], errors='coerce')

data['age'].fillna(data['age'].median(), inplace=True)
data['renta'].fillna(data['renta'].median(), inplace=True)

sns.countplot(data=data, x='segmento')
plt.title('Distribution of Customers Across Different Segments')
plt.show()

sns.boxplot(data=data, x='segmento', y='age')
plt.title('Average Age of Customers in Each Segment')
plt.show()

sns.boxplot(data=data, x='segmento', y='renta')
plt.title('Average Income of Customers in Each Segment')
plt.show()


```
